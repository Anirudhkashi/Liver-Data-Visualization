```python
import pandas as pd
```

```python
data = pd.read_csv("../data/feature_imp.csv", header=None).transpose()
data.columns = ["Feature", "3_months", "6_months", "1_year", "3_years"]
```

```python
data[["3_months", "6_months", "1_year", "3_years"]] = data[["3_months", "6_months", "1_year", "3_years"]].astype(float)
data = data.sort_values(["3_months", "6_months", "1_year", "3_years"], ascending=False).reset_index(drop=True)
data
```

```python
def createGraph(df, i):
    
    max_val = df.loc[:, ["3_months", "6_months", "1_year", "3_years"]].max()
    
    trace1 = go.Bar(
        x=df["Feature"].tolist(),
        y=df["3_months"].tolist(),
        name='3 months',
        marker=dict(
            color='rgb(55, 83, 109)'
        ),
    )
    
    trace2 = go.Bar(
        x=df["Feature"].tolist(),
        y=df["6_months"].tolist(),
        name='6 months',
        marker=dict(
            color='rgb(26, 118, 255)'
        ),
    )
    
    trace3 = go.Bar(
        x=df["Feature"].tolist(),
        y=df["1_year"].tolist(),
        name='1 year',
        marker=dict(
            color='rgb(0, 168, 107)'
        ),
    )
    
    trace4 = go.Bar(
        x=df["Feature"].tolist(),
        y=df["3_years"].tolist(),
        name='3 years',
        marker=dict(
            color='rgb(55, 0, 50)'
        ),
    )
    
    data = [trace1, trace2, trace3, trace4]
    
    layout = go.Layout(
        title='Feature Importance Data',
        xaxis=dict(
            tickfont=dict(
                size=14,
                color='rgb(107, 107, 107)'
            ),
        ),
        yaxis=dict(
            title='Importance',
            titlefont=dict(
                size=16,
                color='rgb(107, 107, 107)'
            ),
            tickfont=dict(
                size=14,
                color='rgb(107, 107, 107)'
            ),
            autorange=False,
            range=[0, 0.045]
        ),
        legend=dict(
            x=0,
            y=1.0,
            bgcolor='rgba(255, 255, 255, 0)',
            bordercolor='rgba(255, 255, 255, 0)'
        ),
        barmode='group',
        bargap=0.15,
        bargroupgap=0.1
    )

    return data, layout
```
