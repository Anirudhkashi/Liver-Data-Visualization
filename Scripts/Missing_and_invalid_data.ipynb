```python
import plotly.plotly as py
import plotly.graph_objs as go
import pandas as pd
```

```python
missing_df = pd.read_csv("../data/missing_data.csv")
invalid_df = pd.read_csv("../data/invalid_data.csv")
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', 3):
    print missing_df.sort_values(by=["Missing"], ascending=False)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', 3):
    print invalid_df.sort_values(by=["Invalid"], ascending=False)
```

```python
# import plotly
# plotly.tools.set_credentials_file(username='anirudhkashi', api_key='v4vvAWcUZ7tKQuPa01IZ')
trace0 = go.Bar(
    x= missing_df["Feature"].tolist(),
    y = missing_df["Missing"].tolist(),
    name='Missing data',
    marker=dict(
        color='rgb(55, 83, 109)',
    )
)

data = [trace0]
py.iplot(data, filename='bar-missing')
```

```python
trace1 = go.Scatter(
    x= missing_df["Feature"].tolist(),
    y = missing_df["Missing"].tolist(),
    mode='markers',
    marker=dict(
        size = missing_df["Missing"].tolist(),
    )
)

data = [trace1]
py.iplot(data, filename='bubble-missing')
```
